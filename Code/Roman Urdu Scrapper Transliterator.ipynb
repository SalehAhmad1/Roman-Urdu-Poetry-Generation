```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import NoSuchElementException
import numpy as np
from tqdm import tqdm
```

```python
#Function to save file
def WriteData(data, filename):
    with open(filename, 'at') as f:
        for line in data:
            f.write(line + '\n')
```

```python
#Reading data previously processed
ProcessedDataOG = None
with open('../Data/Processed_Urdu.txt', 'rt') as f:
    ProcessedDataOG = f.read().splitlines()
ProcessedDataOG
```

```python
ProcessedData = ProcessedDataOG.copy()
```

```python
import string
EnglishAlphabets = string.ascii_lowercase
EnglishAlphabets += string.ascii_uppercase
EnglishAlphabets
```

```python
RomanTransliteratedData = []
for idx,verse in enumerate(ProcessedData):
    Roman = None
    print('Index:', idx)
    print(verse)
    if verse == '' or verse == ' ': #empty line
        Roman = ''
       
    elif verse[0] not in EnglishAlphabets:#check if character is english or not #urdu
        url='https://www.ijunoon.com/transliteration/urdu-to-roman/?type=48202294657'
        options = Options()
        driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
        driver.get(url)

        Xpath = '//*[@id="global"]/div[2]/form/div/textarea'
        driver.find_element(By.XPATH,Xpath).send_keys(verse)

        submit_Xpath='//*[@id="submit"]'
        driver.find_element(By.XPATH,submit_Xpath).click()
        
        translationbox_Xpath='//*[@id="ctl00_inpageResult"]/p'
        Roman = driver.find_element(By.XPATH,translationbox_Xpath).text
    
    else: #roman urdu
        Roman = verse

    WriteData([Roman], '../Data/Transliterated_RomanUrdu.txt')
```
