```python
import numpy as np
```

```python
def WriteData(data, filename):
    with open(filename, 'wt') as f:
        for line in data:
            f.write(line + '\n')
```

```python
#Read The data
UrduDataOG = None
with open('../Data/poetry.txt', 'rt') as f:
    UrduDataOG = f.read().splitlines()
UrduDataOG
```

```python
UrduData = UrduDataOG.copy()
```

# Remove all lines that are not poetry (in poetry 2 consective verses are a must)

# Remove all punctuations from the verses

```python
from typing import FrozenSet, Dict

# Urdu Digits from 0 to 9
URDU_DIGITS: FrozenSet[str] = frozenset("۰ ۱ ۲ ۳ ۴ ۵ ۶ ۷ ۸ ۹".split())

# Urdu Punctuations
URDU_PUNCTUATIONS: FrozenSet[str] = frozenset("؛ ، ٫ ' . ؟ ‘ ! ٴ ۔ ٪".split())
# Urdu Extra Characters
URDU_EXTRA_CHARACTERS: FrozenSet[str] = frozenset(" ؀ ؁ ؂ ؃ ؍ ؎ ؏ ؐ ؑ ؒ ؓ ؔ ؕ ٌ ّ ْ ٓ ٔ ٖ ٗ ٘ ٬".split())

#Save in list all characters in UrduData that are in URDU_PUNCTUATIONS or URDU_EXTRA_CHARACTERS
CharactersToRemove = []
for verse in UrduData:
    for character in verse:
        if character in URDU_PUNCTUATIONS or character in URDU_EXTRA_CHARACTERS:
            CharactersToRemove.append(character)

#Remove the characters in CharactersToRemove
CharactersToRemove = set(CharactersToRemove)
for character in CharactersToRemove:
    UrduData = np.char.replace(UrduData, character, '')
```

# Remove extra new lines after a stanza

```python
#Remove all elements which are space and have a space after and before 
ElementsToRemove = []
LengthOfPoetry = len(UrduData)
for idx,verse in enumerate(UrduData[0:LengthOfPoetry-1]):
    if (verse == '' or verse == ' ') and (UrduData[idx+1] == ''  or UrduData[idx+1] == ' '):
        ElementsToRemove.append(idx)

# remove the elements at the ElementsToRemove
UrduData = np.delete(UrduData, ElementsToRemove)
len(UrduData)
```

```python
WriteData(UrduData, '../Data/Processed_Urdu.txt')
```
